```python
import os 

module_directory = os.path.dirname(os.getcwd())
# define function to find all the files of a given extension in a directory and its subfolders, 
# and return the filename if check_print is True 
def findext(extension, path, check_print): 
    l = [] 
    for root, dirs, files in os.walk(path):
        for filename in files:
            if filename.endswith(extension):
                l.append(os.path.join(root,filename))
                if check_print == True : 
                    print(os.path.splitext(filename)[0])
    return(l)

# first, let's find all the csv files to convert: 
list_csv = findext('csv',module_directory,1) 
```

```python
#define the plotting routine 
import pandas as pd 
import matplotlib.pyplot as plt

def plot_panda(panda_data,filename, plane): 
    print(plane.lower())
    if ('x' and 'y') in plane.lower():
        panda_data.plot(x = 'xh', y = 'yh', marker = 'o', kind = 'scatter', color = 'blue', grid = True,  alpha = 0.8, figsize=(8,8), title ="Orbital Evolution in XY plane (AU) for file {id:s}".format(id = filename))
    elif ('x' and 'z') in plane.lower():
         panda_data.plot(x = 'xh', y = 'zh', marker = '.', kind = 'scatter', color = 'blue', grid = True, alpha = 0.8, figsize=(8,8), title ="Orbital Evolution in XZ plane (AU) for file {id:s}".format(id = filename))
    elif ('y' and 'z') in plane.lower():
         panda_data.plot(x = 'yh', y = 'zh', marker = 'x', kind = 'scatter', color = 'blue', grid = True, alpha = 0.8, figsize=(8,8), title ="Orbital Evolution in YZ plane (AU) for file {id:s}".format(id = filename))
```

```python
# extract data from csv to panda and plot the files accordingly in XY plane 

for file in list_csv: 
    data = pd.read_csv(file, index_col = 0)
    plot_panda(data, file.rsplit('/')[-1].rsplit('.')[0],'XY') 
    
```

```python
# extract data from csv to panda and plot the files accordingly in XZ plane 

for file in list_csv: 
    data = pd.read_csv(file, index_col = 0)
    plot_panda(data, file.rsplit('/')[-1].rsplit('.')[0],'XZ') 
    
    
    
    
```
